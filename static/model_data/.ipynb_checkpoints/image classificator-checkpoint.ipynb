```python
import os
import cv2
```

```python
image_dirs = []
image_class = {}
count = 0

X = []

Y = []

for entry in os.scandir("./cropped"):
    if entry.is_dir():
        image_dirs.append(entry.path)
        
        class_name = entry.path.split("\\")[-1]
        
        print(class_name)
        
        

        for entry_2 in os.scandir(entry.path):
            image = cv2.imread(entry_2.path)
            
            grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            scaled_grayscale_image = cv2.resize(grayscale_image, (400, 400))
            
            X.append(scaled_grayscale_image)
            Y.append(count)
            
        image_class[class_name] = count
        
        count += 1
            

class_list = ['atiku abubakar', 'buhari', 'peter obi', 'tinubu']

class_list
```

```python
X = np.array(X)
Y = np.array(Y)

Y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

plt.imshow(x_train[0], cmap="gray")
print(y_train[0])
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(400, 400)), 
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(4, activation="softmax")
    
])
```

```python
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(x_train, y_train, epochs=22)

test_loss, test_acc = model.evaluate(x_test, y_test)

test_loss, test_acc
```

```python

class_list = ['atiku abubakar', 'buhari', 'peter obi', 'tinubu']

prediction = model.predict(x_test)

p = class_list[np.argmax(prediction[2])]

print(y_test[2])
```
